```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
import ast

df = pd.read_csv('tmdb_5000_movies.csv')
df['new genre'] = df['genres'].apply(ast.literal_eval)

new = []
for i in df['new genre']:
    new = new + i
new1 = []
for j in new:
    new1.append(j["name"])

result = dict(Counter(new1))
x= result.keys()
y = result.values()
plt.bar(x, y)
plt.tick_params(axis='x', labelrotation=45)
plt.xlabel("Genres")
plt.ylabel("Number of movies")
plt.title("Number of Movies per genre")
plt.show()
```
